# 0. An end to end Scikit-learn Workflow

```python
# 1. Get the data ready
import pandas as pd
heart_disease =pd.read_csv("data/heart-disease.csv")
heart_disease
```

```python
#create X(feature metrics)
x = heart_disease.drop("target", axis=1)
# Create Y (labels)
y = heart_disease["target"]
```

```python
# 2. Choose the right estimator/algorithm for the problem
from sklearn.ensemble import RandomForestClassifier
clf =RandomForestClassifier()

#Keep the default hyperparameters
clf.get_params();
```

```python
# 3. Fit the model/algorithm and use it to make predictions on our data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test =train_test_split(x,y, test_size=0.2)
```

```python
clf.fit(x_train, y_train);
```

```python
# make prediction on test data

y_preds =clf.predict(x_test)
```

```python
y_test
```

```python
#4. Evaluating a model on the training and test data
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Improving the model
#Try diff amount of n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimator...")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"Model accuracy on test set: {clf.score(x_test, y_test)*100:.2f}%")
    print("")
```

```python
# 6. Save and load a trained model
import pickle

pickle.dump(clf, open("random_forest.model_1.pkl", "wb"))
```

```python
#upload saved model
loaded_model = pickle.load(open("random_forest.model_1.pkl", "rb"))
loaded_model.score(x_test, y_test)
```

```python
#Make sure data is all numerical
car_sales =pd.read_csv("data/sklearn-data/car-sales-extended.csv")
car_sales.head()
```

```python
len(car_sales)
```

```python
#Split data into X/Y
x = car_sales.drop("Price", axis=1)
y = car_sales["Price"]

#Split data into training and test
x_train, x_test, y_train, y_test =train_test_split(x,y, test_size=0.2)
```

```python
#Build a ML model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
#Get an error above bc ML model doesn't understand strings
#Covert the string colns to floats

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features =["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                   one_hot,
                                   categorical_features)],
                                   remainder="passthrough")
transformed_x = transformer.fit_transform(x)
transformed_x
```

```python
pd.DataFrame(transformed_x)
```

```python
#Using pandas function to turn the same strings into numerical values
dumies = pd.get_dummies (car_sales[["Make", "Colour", "Doors"]])
dumies.head()
```

```python


# Assuming 'transformed_x' is your feature matrix and 'y' is your target values

# Split the data into training and testing sets

x_train, x_test, y_train, y_test = train_test_split(transformed_x, y, test_size=0.2, random_state=42)

# Fit the model to the training data
model.fit(x_train, y_train)
```

```python
# Evaluate the model's performance on the test data
model.score(x_test, y_test)
```
